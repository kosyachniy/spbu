```python
df1 = pd.read_csv('2014/1-Jan14.csv', parse_dates=['Date'], usecols=[0, 2, 3, 4, 5, 6, 7, 8])
df2 = pd.read_csv('2014/2-Feb14.csv', parse_dates=['Date'], usecols=[0, 2, 3, 4, 5, 6, 7, 8])
df = pd.concat([df1, df2])
```

```python
df_week = df.set_index('Date').resample('W')
```

```python
df_cur = df[df.Date == '2014-01-18 00:00:00']
df_cur = df_cur[df_cur.LinkRef == 'AL1470']
df_cur = df_cur[["Flow"]]
```
